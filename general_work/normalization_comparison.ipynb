```python
# Function to plot dataset with AUC and Accuracy on the x-axis
def plot_dataset_revised(name, data):
    metrics = ['Accuracy', 'AUC']
    mnist_scores = [data['MNIST'][metric] for metric in metrics]
    imagenet_scores = [data['Imagenet'][metric] for metric in metrics]

    x = np.arange(len(metrics))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, mnist_scores, width, label='MNIST')
    rects2 = ax.bar(x + width/2, imagenet_scores, width, label='Imagenet')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_xlabel('Metrics')
    ax.set_ylabel('Scores')
    ax.set_title(f'Scores by metric for {name}')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()

    # Attach a text label above each bar in *rects*, displaying its height.
    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(round(height, 2)),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')

    autolabel(rects1)
    autolabel(rects2)

    fig.tight_layout()

    plt.show()

# Plotting all datasets with revised method
for dataset, data in datasets.items():
    plot_dataset_revised(dataset, data)
```
