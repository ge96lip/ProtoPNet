```python
import matplotlib.pyplot as plt
import pandas as pd

# Data
data = {
    'Prototypes per Class': [3, 5, 10],
    'Mean test MCC': [0.9257209069851233, 0.8893951854337182, 0.9157665975917759],
    'Mean test Accuracy': [0.9647435897435898, 0.9471153846153847, 0.9599358974358975],
    'Mean test AUC': [0.9949594564979181, 0.991891299583607, 0.9916173570019724],
    'Training time (minutes)': [58 + 8/60, 55 + 42/60, 49 + 5/60]  # Converting training time to minutes
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot Mean test MCC, Accuracy, AUC
ax1.plot(df['Prototypes per Class'], df['Mean test MCC'], marker='o', label='Mean test MCC')
ax1.plot(df['Prototypes per Class'], df['Mean test Accuracy'], marker='o', label='Mean test Accuracy')
ax1.plot(df['Prototypes per Class'], df['Mean test AUC'], marker='o', label='Mean test AUC')

# Set labels and title
ax1.set_xlabel('Prototypes per Class')
ax1.set_ylabel('Performance Metrics')
ax1.set_title('Performance Comparison of Different Models')
ax1.legend(loc='upper left')


# Show plot
plt.show()
```

```python
import numpy as np

# Data
categories = ['Mean test MCC', 'Mean test Accuracy', 'Mean test AUC']
models = ['3 Prototypes', '5 Prototypes', '8 Prototypes']
values = [
    [0.9257209069851233, 0.9647435897435898, 0.9949594564979181],
    [0.8893951854337182, 0.9471153846153847, 0.991891299583607],
    [0.9157665975917759, 0.9599358974358975, 0.9916173570019724]
]

# Create DataFrame
df = pd.DataFrame(values, columns=categories, index=models)

# Plot
fig, ax = plt.subplots(figsize=(12, 8))

# Define bar width
bar_width = 0.2
x = np.arange(len(categories))

# Plot bars
for i, model in enumerate(models):
    ax.bar(x + i*bar_width, df.loc[model], width=bar_width, label=model)

# Set labels and title
ax.set_xlabel('Performance Metrics')
ax.set_ylabel('Values')
ax.set_title('Performance Comparison of Different Models')
ax.set_xticks(x + bar_width)
ax.set_xticklabels(categories)
ax.legend(title='Models')

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# New data
data_new = {
    'Prototypes per Class': [3, 5, 8],
    'Mean test MCC': [0.5816750507471109, 0.5453352502465185, 0.639368169805752],
    'Mean test Accuracy': [0.8461538461538461, 0.8333333333333334, 0.8653846153846153],
    'Mean test AUC': [0.8398078529657477, 0.8251879699248119, 0.8571428571428573],
    'Training time (minutes)': [5 + 31/60, 6 + 36/60, 5 + 24/60]  # Converting training time to minutes
}

# Create DataFrame
df_new = pd.DataFrame(data_new)

# First Plot: Line plot with dual y-axes
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot Mean test MCC, Accuracy, AUC
ax1.plot(df_new['Prototypes per Class'], df_new['Mean test MCC'], marker='o', label='Mean test MCC')
ax1.plot(df_new['Prototypes per Class'], df_new['Mean test Accuracy'], marker='o', label='Mean test Accuracy')
ax1.plot(df_new['Prototypes per Class'], df_new['Mean test AUC'], marker='o', label='Mean test AUC')

# Set labels and title
ax1.set_xlabel('Prototypes per Class')
ax1.set_ylabel('Performance Metrics')
ax1.set_title('Performance Comparison of Different Models')
ax1.legend(loc='upper left')

# Create a second y-axis to plot training time
ax2 = ax1.twinx()
ax2.plot(df_new['Prototypes per Class'], df_new['Training time (minutes)'], marker='o', color='r', label='Training time (minutes)')
ax2.set_ylabel('Training Time (minutes)')
ax2.legend(loc='upper right')

# Second Plot: Block figure
# Data
categories_new = ['Mean test MCC', 'Mean test Accuracy', 'Mean test AUC']
models_new = ['3 Prototypes', '5 Prototypes', '8 Prototypes']
values_new = [
    [0.5816750507471109, 0.8461538461538461, 0.8398078529657477],
    [0.5453352502465185, 0.8333333333333334, 0.8251879699248119],
    [0.639368169805752, 0.8653846153846153, 0.8571428571428573]
]

# Create DataFrame
df_block = pd.DataFrame(values_new, columns=categories_new, index=models_new)

# Plot
fig_block, ax_block = plt.subplots(figsize=(12, 8))

# Define bar width
bar_width = 0.2
x = np.arange(len(categories_new))

# Plot bars
for i, model in enumerate(models_new):
    ax_block.bar(x + i*bar_width, df_block.loc[model], width=bar_width, label=model)

# Set labels and title
ax_block.set_xlabel('Performance Metrics')
ax_block.set_ylabel('Values')
ax_block.set_title('Performance Comparison of Different Models')
ax_block.set_xticks(x + bar_width)
ax_block.set_xticklabels(categories_new)
ax_block.legend(title='Models')

# Show plots
plt.show()
```
